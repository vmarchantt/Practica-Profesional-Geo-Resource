# Singular Spectrum Analysis (SSA), librería pyts

### Singular Spectrum Analysis

La idea principal de este método es suponer que una serie de tiempo puede ser descompuesta en una suma de diferentes componentes tales como la tendencia, periodos y ruido. SSA puede ser utilizado para hacer suave una serie, reducción de ruido, extraer tendencia y periocidades.

Para más información: Singular Spectrum Analysis for Time Series, N. Golyadina, A Zhigljavsky.

La idea es cargarle las series obtenidas en el scripts resistics_2.ipynb
